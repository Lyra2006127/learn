```python
a = torch.tensor([1, 2, 3])
b = torch.tensor([4, 5, 6])

# 加法
print(a + b)  # tensor([5, 7, 9])

# 乘法（对应位置相乘）
print(a * b)  # tensor([4, 10, 18])

# 矩阵乘法
A = torch.tensor([[1, 2], [3, 4]])
B = torch.tensor([[5, 6], [7, 8]])
print(torch.mm(A, B))  # 矩阵相乘
```
